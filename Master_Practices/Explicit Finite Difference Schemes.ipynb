# **Equaciones en Deriviadas Parciales: Esquema de diferencias finitas explícitas**

**Introducción:**

Los esquemas de diferencias finitas explícitos son un tipo de método numérico utilizado para resolver ecuaciones en derivadas parciales (EDP) mediante la discretización del dominio y la aproximación de las derivadas parciales en función de los valores conocidos en pasos anteriores en el tiempo. En estos esquemas, la solución en un punto de la malla que se calcula directamente a partir de los valores en puntos anteriores en el tiempo, sin necesidad de resolver un sistema de ecuaciones simultáneas.

![Descripción de la imagen](../Images/Esquema1.png)

En general existe una manera de modelar precio de las opciones con una solución exacta. A partir de consideraciones como el Movimiento Geométrico Browniano (GBM) para el precio del subyacente y condiciones de mercado perfecto, con las EDP de Black-Scholes se puede modelar el precio de la opción. 

![Descripción de la imagen](../Images/Esquema2.png)

donde:

- V es el precio de la opción en función del precio del activo subyacente S y el tiempo t.
- σ es la volatilidad del activo subyacente.
- r es la tasa de interés libre de riesgo.

En el caso práctico se puede encontrar la solución al precio de una call por ejemplo, a partir de una distribución acumulada y los demás parámetros.

![Descripción de la imagen](../Images/Esquema3.png)

Se obtiene por tanto una solución exacta al problema que se plantea para el cálculo de la opción. Sin embargo, existen escenarios en los que la valoración de una opción se hace más compleja como por ejemplo el cálculo de una opción americana o considerar condiciones del mercado de manera más realistas. Es bajo estas condiciones que las modelos de métodos numéricos cobran relevancia.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
```

```python
E = 10
T = 0.5
r = 0.05
sigma = 0.2
S1 = 2
S2 = 16
a = np.log(S1 / E)
b = np.log(S2 / E)
M = 80
Mb = math.trunc(M * (b / -a))
h = (-a) / M #Paso del tiempo
N = 50
k = T / N #Paso de espacio
alpha = k * (sigma ** 2) / (2 * h ** 2)
alpha
```

```python
# Definimos los vectores y matrices que usaremos
V = np.zeros((M + Mb + 1, N + 1))
x = np.zeros(M + Mb + 1)

# Aplicamos una transformación
for i in range(M + Mb + 1):
    x[i] = a + (i * h)
    
    
# Calculamos la condición inicial
for i in range(M + Mb + 1):
    V[i, 0] = np.maximum(0, (E - E * np.exp(x[i]))*np.exp(-r * (T)) )
    
    
# Calculamos las condiciones de frontera
for i in range(N + 1):
    V[0, i] = (E - S1) * np.exp(-r * (T - i * k))
    V[-1, i] = 0


# Calculamos el resto de la EDP
for i in range(1, N + 1):
    for j in range(1, M + Mb):
        V[j, i] = alpha * V[j - 1, i - 1] + (1 - 2 * alpha) * V[j, i - 1] + alpha * V[j + 1, i - 1]


#Printamos la evolucion de la edp
S = E*np.exp(x)

# Graficar los resultados
plt.figure(figsize=(10, 6))
for i in range(0, N+1, int(N/5)):
    plt.plot(S, V[:, i], label=f't={i*k:.2f}')
plt.xlabel('Precio del activo subyacente, S')
plt.ylabel('Valor de la opción, V')
plt.legend()
plt.title('Valor de la opción de venta europea usando diferencias finitas')
plt.grid(True)
plt.show()
```

```python
#Mostramos los valores cercanos al precio de ejercicio
for i in range(M-10,M+10,2):
    print(S[i],V[i, N])   
```

En este caso, este print se interpreta de la siguiente manera, por ejemplo:

Cuando el precio del subyacente es de 8.177654339579425 su valor de la opción PUT es de 1.7471020882062165. Y así con el resto de valores.

Ahora en el siguiente apartado lo que vamos a realizar es calcular mediante B-S el precio exacto de la opción de PUT y asi poder hacer una comparativa con el resultado del metodo de Esquema de diferencias finitas explícitas. 

```python
def black_scholes_put(S, E, T, r, sigma):
    d1 = (np.log(S / E) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return E * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

# Parámetros
S1 = 8.177654339579425
S2 = 11.74618943088019
E = 10
T = 0.5
r = 0.05
sigma = 0.2

# Calcular el precio de la opción de venta usando la fórmula de Black-Scholes para S1 y S2
exact_solution_S1 = black_scholes_put(S1, E, T, r, sigma)
exact_solution_S2 = black_scholes_put(S2, E, T, r, sigma)

print("Solución exacta (Black-Scholes) para S1:", exact_solution_S1)
print("Solución exacta (Black-Scholes) para S2:", exact_solution_S2)
```

Esto significa que si el precio del activo subyacente es 8.177, entonces el valor de una opción de venta europea calculado utilizando el modelo de Black-Scholes es 1.6397911185531147. Y, si el precio del activo subyacente es 11.746, entonces el valor de una opción de venta europea calculado utilizando el modelo de Black-Scholes es 0.00010300076473458448. Podemos observar, que los resultado se asemejan a los que hemos obtenido mediante el Esquema de diferencias finitas explícitas. 

**Errores absolutos del esquema de diferencias finitas explícitas**

- El error absoluto es |1.7471020882062165 - 1.6397911185531147| = 0.1073109696531018. (aprox 10^-1)
- El error absoluto es |0.0816357126228286 - 0.06665580404734772| = 0.014979908575480884. (aprox 10^-2)

Estos valores de error absoluto representan los errores absolutos entre las soluciones obtenidas mediante el esquema de diferencias finitas y las soluciones exactas del modelo de Black-Scholes para los precios de activo subyacente S1 = 8.177654339579425 y S2 = 11.74618943088019. Podemos observar que obtenemos errores absolutos pequeños.

**Solución:** reducir el paso de tiempo, por ejemplo: Si M = 100 y N = 80

```python
E = 10
T = 0.5
r = 0.05
sigma = 0.2
S1 = 2
S2 = 16
a = np.log(S1 / E)
b = np.log(S2 / E)
M = 100
Mb = math.trunc(M * (b / -a))
h = (-a) / M
N = 80
k = T / N
alpha = k * (sigma ** 2) / (2 * h ** 2)
alpha
```

```python
# Definimos los vectores y matrices que usaremos
V = np.zeros((M + Mb + 1, N + 1))
x = np.zeros(M + Mb + 1)

# Aplicamos una transformación
for i in range(M + Mb + 1):
    x[i] = a + (i * h)

# Calculamos la condición inicial
for i in range(M + Mb + 1):
    V[i, 0] = np.maximum(0, (E - E * np.exp(x[i])) * np.exp(-r * (T)))

# Calculamos las condiciones de frontera
for i in range(N + 1):
    V[0, i] = (E - S1) * np.exp(-r * (T - i * k))
    V[-1, i] = 0

# Calculamos el resto de la EDP
for i in range(1, N + 1):
    for j in range(1, M + Mb):
        V[j, i] = alpha * V[j - 1, i - 1] + (1 - 2 * alpha) * V[j, i - 1] + alpha * V[j + 1, i - 1]

# Graficar los resultados
S = E * np.exp(x)

plt.figure(figsize=(10, 6))
for i in range(0, N + 1, int(N / 5)):
    plt.plot(S, V[:, i], label=f't={i * k:.2f}')
plt.xlabel('Precio del activo subyacente, S')
plt.ylabel('Valor de la opción, V')
plt.legend()
plt.title('Valor de la opción de venta europea usando diferencias finitas')
plt.grid(True)
plt.show()
```

**Conclusiones:**

 - Incrementar M significa que el tamaño del paso espacial h se reduce, lo cual ganaremos precisión, pero aumentamos el tiempo de cálculo.

 - Al reducir h, el valor de alpha (que depende de h) puede incrementarse si k se mantiene constante.

 - La condición de estabilidad requiere que alpha no exceda un cierto valor. Si se vuelve demasiado grande, el esquema explícito puede volverse inestable.

 - En definitiva, para mantener la estabilidad del esquema explícito, es crucial recalcular k de manera que alpha no exceda el límite permitido, típicamente alpha < 1/2
